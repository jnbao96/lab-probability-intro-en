## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
white = 10
red = 12
black = 8
total = white + red + black
p1 = white/total
p2 = (white/total) * (black/(total - 1))  # without replacement
p3 = (white/total) * (black/total)         #with replacement
p4 = (white/total) * (black/(total - 1))  * (red/(total - 2))  # all different colors

print(f"Taking a white ball out: {p1:.3f}")
print(f"Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one: {p2:.3f}")
print(f"Taking a white ball on first extraction and a black ball out on second extraction with replacement: {p3:.3f}")
print(f"Taking a white ball, then a black ball, then a red ball (all different colors): {p4:.3f}")
```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE
p_cloudy = 0.4
p_rain_given_cloudy = 0.5
p_rain_non_cloudy = 0.2

p_sat_rainy = (p_cloudy * p_rain_given_cloudy) + ((1 - p_cloudy) * p_rain_non_cloudy)
print(f'What is the chance of rain during next Saturday?: {p_sat_rainy:.3f}')
```
